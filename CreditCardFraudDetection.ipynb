```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=0)
```

## CatBoost Classifier

```python
from catboost import CatBoostClassifier
cc = CatBoostClassifier()
cc.fit(X_train,y_train)
```

## RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 20, criterion='entropy', random_state=0 )
rf.fit(X_train,y_train)
```

```python
# CatBoost
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
y_pred_cat = cc.predict(X_test)
cm = confusion_matrix(y_test,y_pred_cat)
cr = classification_report(y_test,y_pred_cat)
print(cm)
print(cr)
accuracy_score(y_test,y_pred_cat)
```

```python
# Random Forest
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
y_pred_rf = rf.predict(X_test)
cm = confusion_matrix(y_test,y_pred_rf)
cr = classification_report(y_test,y_pred_rf)
print(cm)
print(cr)
accuracy_score(y_test,y_pred_rf)
```

## Balancing data by undersampling

```python
# lets try by balancing it 
# 0    284315
# 1       492
majority_class = dataset[dataset['Class'] == 0]
minority_class = dataset[dataset['Class'] == 1]

majority_reducedata = majority_class.sample(n = 473, random_state=42)

new_balance_df = pd.concat([majority_reducedata, minority_class])
new_balance_df = new_balance_df.sample(frac = 1, random_state=42).reset_index(drop=True)
```

```python
class_count_new = new_balance_df["Class"].value_counts()
class_count_new
```

```python
plt.figure(figsize=(6,4))
class_count_new.plot(kind = "bar", color = ["red","green"])
plt.title("Balance Class Distribution")
plt.xlabel("Class")
plt.xticks(rotation = 0)
plt.ylabel("Count")
plt.show()
```

```python
X_new = new_balance_df.iloc[:,:-1].values
y_new = new_balance_df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, train_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 20, criterion='entropy', random_state=0 )
classifier.fit(X_train_new , y_train_new)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
y_pred_new = classifier.predict(X_test_new)
cm1 = confusion_matrix(y_test_new,y_pred_new)
cr1 = classification_report(y_test_new,y_pred_new)
print(cm1)
print(cr1)
accuracy_score(y_test_new,y_pred_new)  
```
